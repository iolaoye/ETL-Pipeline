```python
#Extract the numerical columns from the data
data6_Num = data6.iloc[:,9:] # Select From 11th to end
data6_Num
```

```python
#Drop the numerical column fron the dataset to be left with categorical columns
data6a = data6.drop(['Damage', 'LONGITUDE','LATITUDE', 'Age' ], axis =1)
data6a
```

```python
#Concatenate the Numerical and categorical dataframe 
Fengd_data = pd.concat([data6a_OneHot, data6_Num], axis=1)
Fengd_data
```

```python
Fengd_data.columns
```

```python
Fengd_data = Fengd_data.drop(['index' ], axis =1)
Fengd_data
```

```python
#Shuffle the data
Fengd_data = Fengd_data.sample(frac=1).reset_index(drop=True)
Fengd_data
```

```python
#Building the model
#Split the data into independent (X) and target (y)
X = Fengd_data.iloc[:, :-1]. values
X
```

```python
y = Fengd_data.iloc[:, -1].values
y
```

```python
from sklearn.preprocessing  import LabelEncoder
Ty = LabelEncoder()
y = Ty.fit_transform(y)
y
```

```python
Ty.classes_
```

```python
Ty.inverse_transform([1])
```

```python
#Split the data into train-test
from sklearn.model_selection import train_test_split

#Trian and test split
X_train, X_test,y_train,y_test = train_test_split(X, y, test_size = 0.3, random_state = 234)

```

```python
X_train
```

```python
#Training of the Model
import xgboost as xgb
xgb_model = xgb.XGBClassifier(random_state=234)
xgb_model.fit(X_train,y_train)
```

```python
X_test[0]
```

```python
y_test[0]
```

```python
Pred = xgb_model.predict(X_test)
```

```python
Pred
```

```python
#Model Evaluation
from sklearn.metrics import confusion_matrix
cm=confusion_matrix (y_test,Pred)
cm
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,Pred)*100
```
